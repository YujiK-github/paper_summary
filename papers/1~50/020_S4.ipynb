## 1. どんなもの？
S4, which stands for "Structured State Space Sequence model"の提案。

## 2. 先行研究と比べるとどこが凄い？  
状態空間モデルのパラメータを決めるための計算量やメモリ消費量を落として、使用可能なレベルにしながら、モデルの性能を改善した。既存モデルよりも長いシーケンスの依存関係を捉えられるようになった。  
計算量: $O(N^2L) -> O(N+L)$; メモリ消費量: $O(NL) -> O(N+L)$、$N: \text{state dimension}, L: \text{sequence length}$

## 3. 技術や手法のキモはどこ？


  
* Space State Model  
  計算の簡略化のために$\mathbf{D}=0$とする。
* HiPPO  
  勾配消失、勾配爆発を防ぐために状態遷移行列($A$)をある程度固定化する。-> HiPPO matrixと呼ぶ。
