## 3. 技術や手法のキモはどこ？


  
カウントデータの尤度をnegative Binominal likelihood, 実数値データの尤度を正規分布として、NNの出力値をそれぞれのパラメータとする。  
尤度を最大化させるように学習する。直前の入力値が直後の予測値に対して大きく影響を及ぼすのでアイテムごとに独立したscale factorを導入してscale factorを組み込んだ尤度を使うようにした。また、データの不均衡によってそれぞれの時系列がunderfittingされてしまうことを防ぐために...?  
特徴量にlag dataや類似したアイテムのデータ等を加え、標準化した。

>負の二項分布は、離散確率分布の一つ。確率 p で成功する独立なベルヌーイ試行が繰り返された時の成功回数の分布を表すという意味で二項分布によく似ているが、負の二項分布では試行回数があらかじめ決められておらず、r 回の失敗が起こるまで試行が続けられる。
